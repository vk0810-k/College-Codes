```python
x = np.arange(1,11)
y = 2*x+5

plt.title("Equation of Line")
plt.xlabel("X axis")
plt.ylabel("Y axis")

plt.plot(x,y,'*m'); #circle marker with yellow color
```

```python
x = np.arange(1,11)
y = 2*x+5

plt.title("Equation of Line")
plt.xlabel("X axis")
plt.ylabel("Y axis")

plt.plot(x,y,color = 'black',linestyle = 'dashed',linewidth = 3,
        marker = 'o',markerfacecolor = 'white',markersize = 12); 
```

```python
#x1 is for line 1
x1 = [1,2,3]
y1 = [2,4,1]

plt.plot(x1,y1,label = 'Line 1',color = 'red',linestyle = 'dashed',
         linewidth = 3,marker = 'o',markerfacecolor = 'violet',
         markersize = 12)

#for line 2
x2 = [1,2,3]
y2 = [4,1,3]

plt.plot(x2,y2,label = 'Line 2',color = 'blue',linestyle = 'dashed',
          linewidth = 3,marker = 'o',markerfacecolor = 'yellow',
          markersize = 12)

plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.title("Two line on same graph")
plt.legend()
plt.show()
```

```python
x = [1,2,3,4,5,6]
y = [2,4,1,5,2,6]

#plotting the points
plt.plot(x,y,color = 'black',linestyle = 'dashed',linewidth = 3,
        marker = '<',markerfacecolor = 'yellow',markersize = 12)

#setting x and y range
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.title("Customisation Plot")

plt.show()
```

```python
#compute x and y coordinates of sin and cos curves
x = np.arange(0,3*np.pi,0.1)
y_sin = np.sin(x)
y_cos = np.cos(x)

#set a first subplot grid that has height 2 and width 1
plt.subplot(2,1,1)

#first plot
plt.plot(x,y_sin)
plt.title('SIn Graph')

#set a second subplot grid that has height 2 and width 1
plt.subplot(2,1,2)

#second plot
plt.plot(x,y_cos)
plt.title('Cos Graph')

plt.show()
```

# SCATTER PLOT

```python
x = [5,2,9,4,7]
y = [10,5,8,4,2]

plt.scatter(x,y,color = 'b')

plt.show()
```

```python
x = [1,1.5,2,2.5,3,3.5,3.6]
y = [7.5,8,8.5,9,9.5,10,10.5]

x1 = [8,8.5,9,9.5,10,10.5,11]
y1 = [3,3.5,3.7,4,4.5,5,5.2]

plt.scatter(x,y,label = 'high income low saving',color = 'r')
plt.scatter(x1,y1,label = 'low income high saving',color = 'b')

plt.xlabel('Saving*100')
plt.ylabel('Income*1000')
plt.title("Scatter Plot")
plt.legend()
plt.show()
```

```python
activities = ['eat','sleep','work','play']

slices = [3,7,8,6]

colors = ['r','y','g','b']

plt.pie(slices,labels = activities,colors = colors,
       startangle = 90,shadow = True,explode = (0,0,0.1,0),
        radius = 1.5,autopct = '%1.2f%%') #here 1.2 2 means 2 decimal points
```

# IPL DATASET

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
!pip install seaborn
```

```python
match_data = pd.read_csv('matches.csv')
match_data.head()
```

# 8 most played venue in history of IPL

```python
mpv = match_data['venue'].value_counts()
mpv = mpv[:8]

plt.bar(mpv.index,mpv.values,align = 'center')
plt.title("Most Played Venue")
plt.xlabel("Venue")
plt.ylabel("Name of Stadium")
```

```python
sns.set_style('darkgrid')

mpv = match_data['venue'].value_counts()
mpv = mpv[:8]

plt.figure(figsize = (20,6))

most_played = sns.barplot(mpv.index,mpv.values,alpha = 0.9)


plt.title("Most Played Venue",fontsize = 15)
plt.xlabel("Venue",fontsize = 12)
plt.ylabel("Name of Stadium",fontsize = 12)

most_played.set_xticklabels(rotation =20, labels = mpv.index,fontsize = 12)
```

# Man Of The Match Across 9 season

```python
mom = match_data['player_of_match'].value_counts()
mom = mom[:10]
sns.set_style("darkgrid")
plt.figure(figsize=(20,6))
mom_plot = sns.barplot(mom.index, mom.values, alpha=0.8, palette = 'autumn')
plt.title("Player of the Match", fontsize = 15)
plt.ylabel("Count", fontsize = 15)
plt.xlabel("Name of the Stadium", fontsize = 12)
mom_plot.set_xticklabels(rotation = 90, labels = mom.index, fontsize=15)
plt.show()

```

# matches played across season

```python
plt.figure(figsize = (20, 6))
sns.countplot(x='season', data=match_data,palette = 'summer')
plt.xlabel("Season", fontsize = 20)
plt.show()
```

```python
match_2017 = match_data[match_data['season']==2017]
df = match_2017[match_2017['toss_winner'] == match_2017['winner']]
slices=[len(df), (59-len(df))]
labels=['Yes','No']
plt.pie(slices, labels=labels, startangle=90,
shadow = True, explode=(0, 0.1),
colors=['grey', 'cyan'], autopct='%1.2f%%')
plt.show()
```
